## Introduction
In this report, I would be communicating the insights I was able to get from the wrangling of data from the twitter handle @dog_rates also known as WeRateDogs.

The data consists of the downloaded archive of the handle, image prediction file, of the images in the tweets contained in the archive as well as the data from twitter API which contains the number of retweets and likes/favorites of each tweet.

## Visualzation and Insights

To have a visual, I will load the final result of the wrangling process, named twitter_archive_master.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import tweepy
% matplotlib inline 

df = pd.read_csv('twitter_archive_master.csv')
```

There are 1988 tweets in the dataset, these tweets were examined with 22 items.

Let's take a look at the number of images uploaded with each tweet.

```python
df1 = df.groupby('img_num')['s_rating'].sum()
df1.plot(legend = True, figsize = (14,8));
```

```python
df3 = df.groupby('img_num')['s_rating'].value_counts()
df3.plot(kind = 'pie', legend = True, figsize = (14,8));
```

It was found that above that about 80 percent of the tweets contain only one image. 

Also, in the pie chat, it appears that tweets with only one(1) image are rated higher that those with multiple images.

Now taking a look at how frequent @dog_rates tweets.

```python
#Histogram of number of tweets per day
df2 = (pd.to_datetime(df['timestamp'])
       .dt.floor('30d')
       .value_counts()
       .rename_axis('date')
       .reset_index(name='count'))
df2.hist()
plt.xlabel('Day')
plt.ylabel('tweets')
plt.legend()
```

This show that the highest volume of tweets was within the first three months, where we have an average of 12 tweets per day.

```python
pop_stage = df.dog_stage.value_counts()
pop_stage.plot(kind='bar')
plt.xlabel('dog_stage')
plt.ylabel('population')
plt.xticks(rotation=60);
```

The above chart represents the population of each dog stage in the dataset. 

It shows that the  most popular dog stage in the dataset is the pupper and the least popular is the floofer, neglecting those whose stages were not mentioned.
